```python
#W7A1
def binary_search(arr,n):
  left, right = 0, len(arr) - 1
  mid = (left + right)//2
  while left <= right:
    if arr[mid] == n:
      return mid
    elif arr[mid] < n:
      left == mid + 1
    else:
      right == mid + 1
  return -1
arr = list(map(int, input().split()))
n = int(input())
print (binary_search(arr,n))
```

```python
#W7A2
def count_occ(arr, x):
    a = 0
    for i in range (len(arr)):
        if arr[i] == x:
            a += 1
    return a
arr = list(map(int, input().split()))
n = int(input())
print (count_occ(arr,n))
```

```python
#W7A3
def bubble_sort(arr):
    for i in range(len(arr)):
        swapped = False
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j],arr[j+1]=arr[j+1],arr[j]
                swapped=True
            if not swapped:
                break
a=list(map(int,input().split()))
print(bubble_sort(a))
```

```python
#W7A3
def selection_sort(arr):
    n=len(arr)
    for i in range(n):
        min=i
        for j in range(i+1,n):
            if arr[j]<arr[min]:
                min=j
        arr[min],arr[i]=arr[i],arr[min]
a=list(map(int,input().split()))
print(selection_sort(a))
```

```python
#W7A4
def most_frequent(arr):
    if not arr: return None
    count_dict = {}
    first_occurrence = {} 
    for i, x in enumerate(arr):
        count_dict[x] = count_dict.get(x, 0) + 1
        if x not in first_occurrence:
            first_occurrence[x] = i
    max_count = 0
    res = arr[0]
    for x in count_dict:
        if count_dict[x] > max_count:
            max_count = count_dict[x]
            res = x
        elif count_dict[x] == max_count:
            if first_occurrence[x] < first_occurrence[res]:
                res = x
    return res
```

```python
#W7A5
def count_pairs_with_sum(arr, X):
    count = 0
    freq = {}
    for num in arr:
        target = X - num
        if target in freq:
            count += freq[target]
        freq[num] = freq.get(num, 0) + 1
    return count
```

```python
#w7A6
def flatten_list(nested_list):
    flat_list = []
    for sublist in nested_list:
        for item in sublist:
            flat_list.append(item)
    return flat_list
```

```python
#W7A7
import bisect

def length_of_lis(nums):
    sub = []
    for x in nums:
        idx = bisect.bisect_left(sub, x)
        if idx == len(sub):
            sub.append(x)
        else:
            sub[idx] = x
    return len(sub)
```

```python
#W7A8
def solve_intervals(intervals, queries):
    intervals.sort(key=lambda x: x[1] - x[0])
    results = []
    for q in queries:
        found = False
        for l, r in intervals:
            if l <= q <= r:
                results.append(r - l)
                found = True
                break
        if not found:
            results.append(-1)
    return results
```

```python
#W7A9
import bisect

def count_smaller(nums):
    result = []
    sorted_list = []
    for num in reversed(nums):
        idx = bisect.bisect_left(sorted_list, num)
        result.append(idx)
        bisect.insort(sorted_list, num)
    return result[::-1]
```

```python
#W7A10
import bisect

def max_subarray_sum_mod(nums, m):
    max_mod = 0
    prefix_sum = 0
    sorted_prefixes = [0]
    for num in nums:
        prefix_sum = (prefix_sum + num) % m
        max_mod = max(max_mod, prefix_sum)
        idx = bisect.bisect_right(sorted_prefixes, prefix_sum)
        if idx < len(sorted_prefixes):
            max_mod = max(max_mod, (prefix_sum - sorted_prefixes[idx] + m) % m)
        bisect.insort(sorted_prefixes, prefix_sum)
    return max_mod
```
