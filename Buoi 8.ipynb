```python
#W8A1
pi = 3.14
class Cylinder:
    def __init__(self, radius, height):
        self.radius = radius
        self.height = height
    def dien_tich_be_mat(self):
        return 2 * pi * self.radius * (self.radius + self.height)
    def the_tich(self):
        return pi * self.radius**2 * self.height
cyli = Cylinder(5, 10)
print("Diện tích bề mặt hình trụ:", round(cyli.dien_tich_be_mat(), 2))
print("Thể tích hình trụ:", round(cyli.the_tich(), 2))


```

```python
#W8A2
class Date:
    def __init__(self, day, month, year):
        self.day = day
        self.month = month
        self.year = year
    def nam_nhuan(self):
        if (self.year % 4 == 0 and self.year % 100 != 0) or (self.year % 400 == 0):
            return True
        return False
    def so_ngay_trong_thang(self):
        if self.month == 2:
          return 29 if self.nam_nhuan() else 28
        elif self.month in [4, 6, 9, 11]:
          return 30
        elif self.month in [1, 3, 5, 7, 8, 10, 12]:
          return 31
        else:
          return "INVALID"
    def hop_le(self):
        if self.year < 1 or self.month < 1 or self.month > 12 or self.day < 1 or self.day > self.so_ngay_trong_thang():
            return False
        return True
    def ngay_ke_tiep(self):
        if not self.hop_le():
            return "INVALID"
        if self.day < self.so_ngay_trong_thang():
            return Date(self.day + 1, self.month, self.year)
        elif self.month < 12:
            return Date(1, self.month + 1, self.year)
        else:
            return Date(1, 1, self.year + 1)
a, b, c = map(int, input("Nhập ngày, tháng, năm: ").split())
date = Date(a, b, c)
next_date = date.ngay_ke_tiep()
if next_date == "INVALID":
    print("INVALID")
else:
    print(f"Ngày kế tiếp là: {next_date.day}/{next_date.month}/{next_date.year}")
```

```python
#W8A3
from math import sqrt
class Point2:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    def check_goc_toa_do(self):
        if self.x == 0 and self.y == 0:
            return True
        else:
            return False
    def check_truc_hoanh(self):
        if self.y == 0:
            return True
        else:
            return False
    def check_truc_tung(self):
        if self.x == 0:
           return True
        return False
    def tinh_khoang_cach(self):
        return round(sqrt(self.x ** 2 + self.y ** 2),2)
a, b = map(int,input().split())
point2 = Point2(a, b)
if point2.check_goc_toa_do:
    print("M nam tren goc toa do")
elif point2.check_truc_hoanh:
    print("M nam tren truc hoanh")
elif point2.check_truc_tung:
    print("M nam tren truc tung")
else:
    print("M khong nam tren truc hoanh hay truc tung")
print(point2.tinh_khoang_cach())


            
```

```python
#W8A4
class Calculator:
    def __init__(self, num1, num2):
        self.num1 = num1
        self.num2 = num2

    def add(self):
        return self.num1 + self.num2

    def subtract(self):
        return self.num1 - self.num2

    def product(self):
        return self.num1 * self.num2

    def divide(self):
        if self.num2 == 0:
            return "Không thể chia cho 0"
        return self.num1 / self.num2

    def power(self):
        return self.num1 ** self.num2

    def mod(self):
        if self.num2 == 0:
            return "Không thể chia dư cho 0"
        return self.num1 % self.num2

    def set_numbers(self, num1, num2):
        self.num1 = num1
        self.num2 = num2
        print(f"Đã cập nhật hai số mới: {self.num1} và {self.num2}")

# --- Chương trình chính (Menu) ---
def main():
    print("--- CHƯƠNG TRÌNH MÁY TÍNH BỎ TÚI ---")
    try:
        a = float(input("Nhập số thứ nhất: "))
        b = float(input("Nhập số thứ hai: "))
    except ValueError:
        print("Lỗi: Vui lòng nhập số hợp lệ.")
        return

    # Khởi tạo đối tượng Calculator
    calc = Calculator(a, b)

    while True:
        print("\nCHỌN PHÉP TOÁN:")
        print("1. Cộng (+)")
        print("2. Trừ (-)")
        print("3. Nhân (*)")
        print("4. Chia (/)")
        print("5. Lũy thừa (^)")
        print("6. Chia lấy dư (%)")
        print("7. Nhập lại 2 số mới")
        
        choice = input("Nhập lựa chọn của bạn (1-7): ")

        result = None
        
        if choice == '1':
            result = calc.add()
            op = "+"
        elif choice == '2':
            result = calc.subtract()
            op = "-"
        elif choice == '3':
            result = calc.product()
            op = "*"
        elif choice == '4':
            result = calc.divide()
            op = "/"
        elif choice == '5':
            result = calc.power()
            op = "^"
        elif choice == '6':
            result = calc.mod()
            op = "%"
        elif choice == '7':
            try:
                new_a = float(input("Nhập số thứ nhất mới: "))
                new_b = float(input("Nhập số thứ hai mới: "))
                calc.set_numbers(new_a, new_b)
                continue # Quay lại đầu vòng lặp để chọn phép tính ngay
            except ValueError:
                print("Lỗi: Vui lòng nhập số.")
                continue
        else:
            print("Lựa chọn không hợp lệ!")
            continue

        # In kết quả nếu không phải là chọn update số
        if choice in ['1', '2', '3', '4', '5', '6']:
            print(f"Kết quả: {calc.num1} {op} {calc.num2} = {result}")

        # Kiểm tra thoát chương trình đúng theo yêu cầu đề bài
        exit_check = input("\nBạn có muốn thoát? (Go 'yes', 'YES' hoặc 'Yes' de thoat): ")
        if exit_check in ['yes', 'YES', 'Yes']:
            print("Đã thoát chương trình.")
            break

if __name__ == "__main__":
    main()
```
