```python
#W10A2
try:
    line = input().split()
    my_list = [int(x) for x in line]
    n = len(my_list)
    print(my_list[n]) 
except IndexError:
    print("Loi truy cap chi so")
```

```python
#W10A3
def is_happy(n):
    seen = set()
    while n != 1 and n not in seen:
        seen.add(n)
        n = sum(int(d)**2 for d in str(n))
    return n == 1
try:
    val = input()
    n = int(val)
    if is_happy(n):
        print("YES")
    else:
        print("NO")
except ValueError:
    print("Loi nhap lieu")
```

```python
#W10A5
raw_list = input().split()
int_list = []
for item in raw_list:
    try:
        val = int(item)
        int_list.append(val)
    except ValueError:
        print(f"Canh bao: '{item}' khong phai la so nguyen")
print("Danh sach da chuyen:", int_list)
```

```python
#W10A6
import math

try:
    x = float(input())
    if x < 0:
        raise ValueError
    print(f"{math.sqrt(x):.2f}")
except ValueError:
    print("So am")
```
